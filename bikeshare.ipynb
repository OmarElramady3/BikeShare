```python
import time
import pandas as pd
import numpy as np

CITY_DATA = { 'chicago':'chicago.csv',
              'new york city':'new_york_city.csv',
              'washington':'washington.csv' }

     
cities = {1:'chicago' , 2:'new york city' , 3:'washington'}
months = ['January','February','March','April','May','June']
days = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
filters = ['month','day','both','not at all']
def get_filters():
    Filter, city ,month ,day = '0' ,0 ,'all' ,'all'
    """
    Asks user to specify a city, month, and day to analyze.

    Returns:
        (str) city - name of the city to analyze
        (str) month - name of the month to filter by, or "all" to apply no month filter
        (str) day - name of the day of week to filter by, or "all" to apply no day filter
    """
    print('Hello! Let\'s explore some US bikeshare data!')
    # TO DO: get user input for city (chicago, new york city, washington). HINT: Use a while loop to handle invalid inputs
    while city not in range(1,4):
        city = int(input("Choose a city \nEnter 1 for chicago ,2 for new york city and 3 for washington: "))
    city = cities[city]
    
    # TO DO: get user input for month (all, january, february, ... , june)
    # TO DO: get user input for day of week (all, monday, tuesday, ... sunday)

    while Filter not in filters:
        if Filter != '0':
            print('Invalid Value...')
        Filter = input("Would you like to filter the data by month, day, both, or not at all? ").lower()
    
    if Filter == 'month':
        if month != 'all':
            print('Invalid Value...')
        while month not in months:
            month = str(input("Choose a month \nEnter the month like(January ...,June) : ")).title()
    
    elif Filter == 'day':
        if day != 'all':
            print('Invalid Value...')
        while day not in days:
            day = str(input("Choose a day \nEnter a day like(Monday, Tuesday...) : ")).title()
    
    elif Filter == 'both':
        if month != 'all' or day != 'all':
            print('Invalid Value...')
        while month not in months:
            month = str(input("Choose a month \nEnter the month like(January ...,June) : ")).title()
        
        while day not in days:
            day = str(input("Choose a day \nEnter a day like(Monday, Tuesday...) : ")).title()

    print('-'*40)
    return city, month, day
def load_data(city, month, day):
    """
    Loads data for the specified city and filters by month and day if applicable.

    Args:
        (str) city - name of the city to analyze
        (str) month - name of the month to filter by, or "all" to apply no month filter
        (str) day - name of the day of week to filter by, or "all" to apply no day filter
    Returns:
        df - Pandas DataFrame containing city data filtered by month and day
    """
    df = pd.read_csv(CITY_DATA[city])
    df['Start Time'] = pd.to_datetime(df['Start Time'])
    df['month'] = df['Start Time'].dt.month
    df['day of week'] = df['Start Time'].dt.day_of_week
    df['hour'] = df['Start Time'].dt.hour
    if month != 'all':
        month = months.index(month) +1
        df = df[df['month']==month]
    if day != 'all':
        day = days.index(day) 
        df=df[df['day of week']==day]
    return df
def time_stats(df):
    """Displays statistics on the most frequent times of travel."""
    print('\nCalculating The Most Frequent Times of Travel...\n')
    start_time = time.time()
    # TO DO: display the most common month
    print("the most common month :\n\t",months[df['month'].mode()[0]-1])

    # TO DO: display the most common day of week
    print("the most common day of week :\n\t",days[df['day of week'].mode()[0]])

    # TO DO: display the most common start hour
    print('the most common start hour :\n\t',df['hour'].mode()[0])


    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)
def station_stats(df):
    """Displays statistics on the most popular stations and trip."""

    print('\nCalculating The Most Popular Stations and Trip...\n')
    start_time = time.time()

    # TO DO: display most commonly used start station
    print('the most commonly used start station: \n\t',df['Start Station'].mode()[0])

    # TO DO: display most commonly used end station
    print('the most commonly used end station: \n\t',df['End Station'].mode()[0])

    # TO DO: display most frequent combination of start station and end station trip
    # size() function is to know how many 'end stations' in each 'start station'...
    X = df.groupby(['Start Station', 'End Station']).size()
    most_frequent = X.sort_values(ascending=False).head(1)
    
    print('the most frequent combination of start station and end station trip :\n\t',most_frequent)
    
    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)
    
def trip_duration_stats(df):
    """Displays statistics on the total and average trip duration."""

    print('\nCalculating Trip Duration...\n')
    start_time = time.time()

    # TO DO: display total travel time
    print("total travel time :\n\t",df['Trip Duration'].sum())

    # TO DO: display mean travel time
    print("mean travel time :\n\t",df['Trip Duration'].mean())

    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)
def user_stats(df,city):
    """Displays statistics on bikeshare users."""

    print('\nCalculating User Stats...\n')
    start_time = time.time()
    if city == 'washington':
        print("there is no Gender informations")
        print("there is no date of birth informarions")
    else:
        # TO DO: Display counts of user types
        print('counts of user types :\n\t',df['User Type'].value_counts())

        # TO DO: Display counts of gender
        print('counts of Gender :\n\t',df['Gender'].value_counts())

        # TO DO: Display earliest, most recent, and most common year of birth
        print('the most common year of birth :\n\t',df['Birth Year'].mode()[0])
        print('the earliest year of birth :\n\t',df['Birth Year'].min())
        print('the most recent year of birth :\n\t',df['Birth Year'].max())


    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)
def main():
    while True:
        show = 0
        city, month, day = get_filters()
        df = load_data(city, month, day)

        time_stats(df)
        station_stats(df)
        trip_duration_stats(df)
        user_stats(df,city)
        
        while show not in ['yes','no']:
            show = str(input('would you like to see the first 5 row of data? Enter yes or no.\n'))
        
        if show.lower() == 'yes':
            print(df.head())

        restart = input('\nWould you like to restart? Enter yes or no.\n')
        if restart.lower() != 'yes':
            break
        


if __name__ == "__main__":
	main()
```
